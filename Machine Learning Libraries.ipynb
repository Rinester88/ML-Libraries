## 3. a. In the column "mfr", replace the column "K" as "Kellogg's", "G" as "Nestle" and all other values as "Other Brands"

```python
data['mfr'].replace({'K':"Kellogg's",'G':'Nestle'},inplace=True) 
# It used to replace the values of 'mfr' column into the values
```

```python
data # show the data 
```

```python
for i in data['mfr']: # It Iterating the all values of column 'mfr'
    # And its Checking the value same as kellog's and nestle or not in the data
    if i!="Kellogg's" and i!="Nestle":

        data['mfr'].replace({i:'Other Brands'},inplace=True)
        # And if the values are not same, change it into the Other Brands
```

```python
data
```

## 8. Apply MinMaxScaler() and StandardScaler() to the following features: calories, protien, fat, sodium, fiber, carbo, sugars.

```python
# Import MinMaxScaler and StandardScaler from sklearn.preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

 - StandardScaler follows Standard Normal Distribution. Therefore, it makes mean = 0 and scales the data to unit variance. 
 - MinMaxScaler scales all the data features in the range [0, 1] or else in the range [-1, 1] if there are negative values in the dataset.

```python
trans = MinMaxScaler()
# Creating a dataframe of these features: 'calories', 'protein','fat','sodium','fiber','carbo','sugars'
# and transforming all value of these all features using MinMaxScaler
data[['calories', 'protein','fat','sodium','fiber','carbo','sugars']] = trans.fit_transform(data[['calories', 'protein','fat','sodium','fiber','carbo','sugars']])
```

```python
data
```

```python
trans1 = StandardScaler()
# Creating a dataframe of these features: 'calories', 'protein','fat','sodium','fiber','carbo','sugars'
# and transforming all value of these all features using StandardScaler
data[['calories', 'protein','fat','sodium','fiber','carbo','sugars']]= trans1.fit_transform(data[['calories', 'protein','fat','sodium','fiber','carbo','sugars']])
```

```python
data
```
